```python
import pandas as pd
```

Preparing the columns of cities

```python
df = pd.read_csv('laligua_clubs.csv')
df['Club'] = df['Club'].str.replace('\[b\]', '', regex=True)
df['Club'] = df['Club'].str.replace('\[a\]', '', regex=True)
df.to_csv('laligua_clubs.csv', index=False)
```

```python
df = pd.read_csv('ligue1_clubs.csv')
df['Club Name'] = df['Club Name'].str.replace('^\d+\.\s', '', regex=True)
df.to_csv('ligue1_clubs.csv', index=False)
```

```python
df = pd.read_csv('premier_league_clubs.csv')
df['Club'] = df['Club'].str.replace('\[n 2\]', '', regex=True)
df.to_csv('premier_league_clubs.csv', index=False)
```

Grouping the clubs and their academys

```python
bundesliga_transfer = pd.read_csv('/home/ayknes/football_academy_analytics/data/transfer/1-bundesliga.csv')
seriea_transfer = pd.read_csv('/home/ayknes/football_academy_analytics/data/transfer/serie-a.csv')
laligua_transfer = pd.read_csv('/home/ayknes/football_academy_analytics/data/transfer/primera-division.csv')
ligue1_transfer = pd.read_csv('/home/ayknes/football_academy_analytics/data/transfer/ligue-1.csv')
premiere_league_transfer = pd.read_csv('/home/ayknes/football_academy_analytics/data/transfer/premier-league.csv')
top5_ligues = pd.concat([bundesliga_transfer, seriea_transfer, laligua_transfer, ligue1_transfer, premiere_league_transfer])
```

```python
top5_ligues['club_involved_name'] = top5_ligues['club_involved_name'].astype(str)

pattern = r'U18|U19|U21|U23|B$|Res\.'  # the regex pattern
academies = top5_ligues[top5_ligues['club_involved_name'].str.contains(pattern)]
academies['club_involved_name'].value_counts()
```

```python
# import the required library
import re

# create a function to extract the base club name
def extract_base_club_name(name):
    pattern = r'U18|U19|U21|U23|B$|Res\.'  # the regex pattern
    match = re.search(pattern, name)
    if match:
        return name[:match.start()].strip()
    else:
        return name

# apply the function to the 'club_involved_name' column to create a new column 'base_club_name'
top5_ligues['base_club_name'] = top5_ligues['club_involved_name'].apply(extract_base_club_name)

# group by the new column
grouped = top5_ligues.groupby('base_club_name')
# Save the DataFrame to a CSV file
top5_ligues.to_csv('top5_ligues_transfer.csv', index=False)
```
