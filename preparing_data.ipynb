```python
# import the required library
import re

# create a function to extract the base club name
def extract_base_club_name(name):
    pattern = r'U18|U19|U21|U23|B$|Res\.'  # the regex pattern
    match = re.search(pattern, name)
    if match:
        return name[:match.start()].strip()
    else:
        return name

# apply the function to the 'club_involved_name' column to create a new column 'base_club_name'
top_ligues['base_club_name'] = top_ligues['club_involved_name'].apply(extract_base_club_name)

# group by the new column
grouped = top_ligues.groupby('base_club_name')
# Save the DataFrame to a CSV file
top_ligues.to_csv('top_ligues_transfer.csv', index=False)
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

df = pd.read_csv('clubs_ligues_22_23.csv')

geolocator = Nominatim(user_agent="myGeocoder")

# Apply a rate limiter to respect API usage limits
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

# Create location column
df['location'] = df['Location'].apply(geocode)

# Create longitude, latitude and altitude from location column (returns tuple)
df['point'] = df['location'].apply(lambda loc: tuple(loc.point) if loc else None)

# Split point column into latitude, longitude and altitude columns
df[['latitude', 'longitude', 'altitude']] = pd.DataFrame(df['point'].tolist(), index=df.index)

# Drop unnecessary columns
df = df.drop(['location', 'point', 'altitude'], axis=1)
```
