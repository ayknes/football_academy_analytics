```python
bundesliga_transfer = pd.read_csv('/home/ayknes/football_academy_analytics/data/transfer/1-bundesliga.csv')
seriea_transfer = pd.read_csv('/home/ayknes/football_academy_analytics/data/transfer/serie-a.csv')
laligua_transfer = pd.read_csv('/home/ayknes/football_academy_analytics/data/transfer/primera-division.csv')
ligue1_transfer = pd.read_csv('/home/ayknes/football_academy_analytics/data/transfer/ligue-1.csv')
premiere_league_transfer = pd.read_csv('/home/ayknes/football_academy_analytics/data/transfer/premier-league.csv')
eredivisie_transfer = pd.read_csv('/home/ayknes/football_academy_analytics/data/transfer/eredivisie.csv')
liga_nos_transfer = pd.read_csv('/home/ayknes/football_academy_analytics/data/transfer/liga-nos.csv')
top_ligues = pd.concat([bundesliga_transfer, seriea_transfer, laligua_transfer, ligue1_transfer, premiere_league_transfer, eredivisie_transfer, liga_nos_transfer])
```

```python
top_ligues['club_involved_name'] = top_ligues['club_involved_name'].astype(str)

pattern = r'U18|U19|U21|U23|B$|Res\.'  # the regex pattern
academies = top_ligues[top_ligues['club_involved_name'].str.contains(pattern)]
academies['club_involved_name'].value_counts()
```

```python
# import the required library
import re

# create a function to extract the base club name
def extract_base_club_name(name):
    pattern = r'U18|U19|U21|U23|B$|Res\.'  # the regex pattern
    match = re.search(pattern, name)
    if match:
        return name[:match.start()].strip()
    else:
        return name

# apply the function to the 'club_involved_name' column to create a new column 'base_club_name'
top_ligues['base_club_name'] = top_ligues['club_involved_name'].apply(extract_base_club_name)

# group by the new column
grouped = top_ligues.groupby('base_club_name')
# Save the DataFrame to a CSV file
top_ligues.to_csv('top_ligues_transfer.csv', index=False)
```

Merge the two dataframes

```python
clubs_ligues = pd.read_csv('clubs_ligues.csv')
top_ligues_transfer = pd.read_csv('top_ligues_transfer.csv')
clubs_22_23 =top_ligues_transfer[top_ligues_transfer['year'] == 2022]
clubs_22_23 = clubs_22_23.drop_duplicates(subset=['club_name'])



df_merged = pd.merge(clubs_ligues, clubs_22_23[['club_name']], left_on='Club', right_on='club_name', how='left')
df_notnull = df_merged[df_merged['club_name'].notnull()]
df_isnull = df_merged[df_merged['club_name'].isnull()]
df_merged
```

```python
import numpy as np

def fix_null_clubs(row):
    if pd.isnull(row['club_name']):
        club_name = row['Club']
        matches = clubs_22_23['club_name'].str.contains(club_name, case=False, na=False)
        matching_clubs = clubs_22_23.loc[matches, 'club_name']
        if not matching_clubs.empty:
            row['club_name'] = matching_clubs.iloc[0]
    return row

df_merged_fixed = df_merged.apply(fix_null_clubs, axis=1)
df_merged_fixed[df_merged_fixed['club_name'].isnull()]
```
